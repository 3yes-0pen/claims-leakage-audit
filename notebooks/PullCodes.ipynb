```python
import pandas as pd

df = pd.read_csv("procedures.csv")

print("=== PROCEDURE CODES (SNOMED → CPT) ===")
proc_codes = (
    df[["CODE", "DESCRIPTION"]]
    .drop_duplicates()
    .dropna(subset=["CODE"])
    .sort_values("CODE")
)
for _, row in proc_codes.iterrows():
    print(f"{int(row['CODE'])} — {row['DESCRIPTION']}")

print("\n=== REASON CODES (SNOMED → ICD-10) ===")
reason_codes = (
    df[["REASONCODE", "REASONDESCRIPTION"]]
    .drop_duplicates()
    .dropna(subset=["REASONCODE"])
    .sort_values("REASONCODE")
)
for _, row in reason_codes.iterrows():
    print(f"{int(row['REASONCODE'])} — {row['REASONDESCRIPTION']}")
```
